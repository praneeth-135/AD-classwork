```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
iris_data=sns.load_dataset('iris')
df=sns.load_dataset('taxis')
```

```python
sns.scatterplot(x=iris_data['sepal_length'],y=iris_data['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris_data['sepal_length'],y=iris_data['petal_length'],hue=iris_data['species'])
plt.show()
```

```python
sns.scatterplot(x=iris_data['sepal_length'],y=iris_data['petal_length'],hue=iris_data['species'],palette=['r','g','y'])
plt.show()
```

```python
sns.stripplot(x=iris_data['sepal_length'],y=iris_data['petal_length'],hue=iris_data['species'],palette=['r','g','y'])
plt.show()
```
